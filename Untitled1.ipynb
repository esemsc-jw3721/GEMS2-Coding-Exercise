```python
# Define the function to compute mobility f1(S1)
def compute_mobility(S1, S1i, S2r, kr1_max, kr2_max, mu1, mu2, a, b, epsilon=1e-8):
    """
    Computes the mobility function f1(S1) with improved numerical stability.

    Parameters:
    S1 (array): Saturation values.
    S1i (float): Initial saturation.
    S2r (float): Residual saturation of the second phase.
    kr1_max (float): Maximum relative permeability of phase 1.
    kr2_max (float): Maximum relative permeability of phase 2.
    mu1 (float): Viscosity of phase 1.
    mu2 (float): Viscosity of phase 2.
    a (float): Exponent for kr1 calculation.
    b (float): Exponent for kr2 calculation.
    epsilon (float): Small number to prevent numerical issues.

    Returns:
    array: Mobility function values f1(S1).
    """
    # Ensure S1 is within physical limits to prevent unphysical values
    S1 = np.clip(S1, S1i, 1 - S2r)
    eps = 1e-10

    # Compute effective saturation Se, ensuring stability
    Se = np.clip((S1 - S1i) / (1 - S2r - S1i + eps), 0, 1)

    # Compute relative permeabilities with safeguards against numerical issues
    kr1 = kr1_max * np.maximum(Se, eps) ** a  # Avoid zero in exponentiation
    kr2 = kr2_max * np.maximum(1 - Se, eps) ** b  # Avoid zero in exponentiation

    # Compute mobility function f1(S1), avoiding divide-by-zero errors
    f1 = 1 / (1 + (mu1 * kr2) / (mu2 * np.maximum(kr1, epsilon)))  # Prevent zero division

    return f1

# Define saturation range
S1 = np.linspace(0, 1, 100)  # Common range for all cases

# Case 1: CO2 injection into an aquifer
f1_case1 = compute_mobility(S1, S1i=0, S2r=0.2, kr1_max=1, kr2_max=1, mu1=0.1, mu2=1, a=1.2, b=3)

# Case 2: Water displaces hydrogen in a storage site
f1_case2 = compute_mobility(S1, S1i=0.2, S2r=0.5, kr1_max=0.15, kr2_max=1, mu1=1, mu2=0.1, a=4, b=1.5)

# Case 3: Water displacing CO2 in a depleted oil-field
f1_case3 = compute_mobility(S1, S1i=0.2, S2r=0.15, kr1_max=0.4, kr2_max=0.8, mu1=0.4, mu2=0.1, a=8, b=3)

# Plot all cases on the same graph
plt.figure(figsize=(8, 6))
plt.plot(S1, f1_case1, label="Case 1: CO2 Injection", linestyle="--", color="blue")
plt.plot(S1, f1_case2, label="Case 2: Water Displaces Hydrogen", linestyle="-.", color="red")
plt.plot(S1, f1_case3, label="Case 3: Water Displacing CO2", linestyle="-", color="green")

# Labels and title
plt.xlabel("Saturation (S1)")
plt.ylabel("Mobility (f1)")
plt.title("Mobility vs. Saturation for Different Cases")
plt.legend()
plt.grid(True)

# Show the plot

plt.show()
```

```python
def compute_df1_dS1(S1, S1i, S2r, kr1_max, kr2_max, mu1, mu2, a, b, delta_S=0.001):
    f1_plus = compute_mobility(S1 + delta_S/2, S1i, S2r, kr1_max, kr2_max, mu1, mu2, a, b)
    f1_minus = compute_mobility(S1 - delta_S/2, S1i, S2r, kr1_max, kr2_max, mu1, mu2, a, b)
    df1_dS1 = (f1_plus - f1_minus) / delta_S
    return df1_dS1

# Define saturation range
S1_values = np.linspace(0, 1, 100)

# Compute df1/dS1 for different cases
df1_case1 = compute_df1_dS1(S1_values, S1i=0, S2r=0.2, kr1_max=1, kr2_max=1, mu1=0.1, mu2=1, a=1.2, b=3)
df1_case2 = compute_df1_dS1(S1_values, S1i=0.2, S2r=0.5, kr1_max=0.15, kr2_max=1, mu1=1, mu2=0.1, a=4, b=1.5)
df1_case3 = compute_df1_dS1(S1_values, S1i=0.2, S2r=0.15, kr1_max=0.4, kr2_max=0.8, mu1=0.4, mu2=0.1, a=8, b=3)

# Plot the derivatives
plt.figure(figsize=(8, 6))
# plt.plot(S1_values, df1_case1, label="Case 1: CO2 Injection", linestyle="--", color="blue")
plt.plot(S1_values, df1_case2, label="Case 2: Water Displaces Hydrogen", linestyle="-.", color="red")
# plt.plot(S1_values, df1_case3, label="Case 3: Water Displacing CO2", linestyle="-", color="green")

# Labels and title
plt.xlabel("Saturation (S1)")
plt.ylabel("df1/dS1")
plt.title("Derivative of Mobility Function vs. Saturation")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Function to find intersection ignoring zero values
def find_valid_intersection(S1, y_values, df1_values, threshold=1e-3):
    """
    Finds the first intersection where y_values = df1_values, excluding near-zero values.

    Parameters:
    S1 (array): Saturation values.
    y_values (array): Normalized mobility values.
    df1_values (array): Derivative of mobility values.
    threshold (float): Minimum value to consider valid intersection.

    Returns:
    tuple: (S1_shock, y_shock) intersection point.
    """
    for i in range(len(S1) - 1):
        if (y_values[i] > threshold and df1_values[i] > threshold) and (
            (y_values[i] - df1_values[i]) * (y_values[i + 1] - df1_values[i + 1]) < 0
        ):
            # Linear interpolation for more accurate intersection
            S1_shock = S1[i] + (S1[i + 1] - S1[i]) * (df1_values[i] - y_values[i]) / (
                (y_values[i + 1] - y_values[i]) - (df1_values[i + 1] - df1_values[i])
            )
            y_shock = y_values[i] + (y_values[i + 1] - y_values[i]) * (S1_shock - S1[i]) / (S1[i + 1] - S1[i])
            return S1_shock, y_shock
    return None, None

# Compute refined intersection points ignoring zero values
S1_shock_case3_valid, y_shock_case3_valid = find_valid_intersection(S1_case3, y_case3_filtered, df1_case3_filtered)

# Plot Case 3 (Water Displacing CO2) with valid intersection point
plt.figure(figsize=(8, 6))
plt.plot(S1_case3, y_case3_filtered, linestyle="-", color="green", label="y = f1(S1) / (S1 - S1i)")
plt.plot(S1_case3, df1_case3_filtered, linestyle="--", color="purple", label="df1/dS1")
if S1_shock_case3_valid is not None:
    plt.scatter(S1_shock_case3_valid, y_shock_case3_valid, color="black", zorder=3, label="Valid Intersection")
plt.xlabel("Saturation (S1)")
plt.ylabel("Value")
plt.title("Case 3: Water Displacing CO2 - Normalized Mobility & Its Derivative (S1 < 1 - S2r)")
plt.legend()
plt.grid(True)
plt.show()

# Compute intersection points
S1_shock_case1, y_shock_case1 = find_intersection(S1_case1, y_case1_filtered, df1_case1_filtered)
S1_shock_case2, y_shock_case2 = find_intersection(S1_case2, y_case2_filtered, df1_case2_filtered)
S1_shock_case3, y_shock_case3 = find_intersection(S1_case3, y_case3_filtered, df1_case3_filtered)

# Plot Case 1 (CO2 Injection) with intersection point
plt.figure(figsize=(8, 6))
plt.plot(S1_case1, y_case1_filtered, linestyle="--", color="blue", label="y = f1(S1) / (S1 - S1i)")
plt.plot(S1_case1, df1_case1_filtered, linestyle="-.", color="red", label="df1/dS1")
if S1_shock_case1 is not None:
    plt.scatter(S1_shock_case1, y_shock_case1, color="black", zorder=3, label="Intersection")
plt.xlabel("Saturation (S1)")
plt.ylabel("Value")
plt.title("Case 1: CO2 Injection - Normalized Mobility & Its Derivative (S1 < 1 - S2r)")
plt.legend()
plt.grid(True)
plt.show()



```
